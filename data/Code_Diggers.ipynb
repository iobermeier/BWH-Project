```python
# https://www.kaggle.com/datasets/joebeachcapital/food-waste- Download the data

import pandas as pd

# Load the CSV file into a DataFrame
food_waste_data = pd.read_csv('Food Waste data and research - by country.csv')

# Print the first few rows of the DataFrame to view the data
print(food_waste_data.head())
```

```python
# First page- showing facts and some graphs about US - can make this better

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# Assuming the dataset is stored in a variable named 'food_waste_data'

# Filter data for the United States
us_data = food_waste_data[food_waste_data['Country'] == 'United States of America']

# Filter data for other countries
other_countries_data = food_waste_data[food_waste_data['Country'] != 'United States of America']

# Comparison Analysis
us_combined_waste = us_data['combined figures (kg/capita/year)'].values[0]
other_countries_mean_waste = other_countries_data['combined figures (kg/capita/year)'].mean()

# Visualization
plt.figure(figsize=(10, 6))

# Bar chart for comparison
plt.bar(['United States', 'Other Countries'], [us_combined_waste, other_countries_mean_waste], color=['blue', 'green'])
plt.title('Comparison of Food Waste Generation (kg/capita/year)')
plt.ylabel('Food Waste (kg/capita/year)')
plt.show()
```

```python
from difflib import get_close_matches

def is_disposable(item):
    disposables = {
        'paper': ['Paper towels', 'Napkins', 'Tissues', 'Paper plates', 'Paper cups'],
        'plastic': ['Plastic cutlery', 'Drinking straws', 'Plastic plates', 'Plastic cups',
                    'Plastic bags', 'Plastic wrap'],
        'foam': ['Styrofoam cups', 'Styrofoam plates', 'Styrofoam take-out containers'],
        'food containers': ['Take-out containers', 'Aluminum foil containers', 'Plastic clamshell containers'],
        'personal care': ['Disposable razors', 'Disposable wipes', 'Cotton swabs', 'Menstrual products', 'Diapers'],
        'food packaging': ['Plastic wrap', 'Plastic bags', 'Plastic clamshell containers', 'Single-serve food packaging'],
        'miscellaneous': ['Disposable gloves', 'Disposable masks', 'Disposable cameras', 'Disposable batteries'],
        'glass': ['Disposable glassware', 'Disposable glass containers'],
        'metal': ['Disposable aluminum pans', 'Disposable metal cutlery'],
        'compost': ['Fruit peels', 'Vegetable scraps', 'Coffee grounds', 'Tea bags', 'Paper napkins',
                    'Fruits', 'Vegetables', 'Bread', 'Rice', 'Pasta', 'Eggs', 'Meat', 'Fish', 'Dairy products']
    }

    for category, items in disposables.items():
        for disposable_item in items:
            if item.lower() == disposable_item.lower():
                return f"{item} is disposable ({category})."

    # If the input item is not found directly, try to find similar items
    suggestions = []
    for items in disposables.values():
        suggestions.extend(items)

    closest_matches = get_close_matches(item.lower(), suggestions, n=3, cutoff=0.6)
    if closest_matches:
        suggestion_str = ', '.join(closest_matches)
        return f"{item} is not disposable, but similar items include: {suggestion_str}"

    return f"{item} is not disposable."

# Test the function
item = input("Enter the item: ")
print(is_disposable(item))
```
